# Remote Work and Mental Health

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Loading the dataset
df = pd.read_csv("Impact_of_Remote_Work_on_Mental_Health.csv")

# Display first few rows of the datasets
df.head(10)
```

```python
# Data overview

df.info()
```

```python
#get statistical summary of numerical columns
df.describe()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())


```

```python
# Define a function to create a bar chart for satisfaction by region
def plot_satisfaction_by_region(data):
    """
    Plots a bar chart comparing satisfaction with remote work across regions.
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Region', hue='Satisfaction_with_Remote_Work', data=data)
    plt.title("Satisfaction with Remote Work by Region")
    plt.xlabel("Region")
    plt.ylabel("Count")
    plt.legend(title="Satisfaction Level")
    plt.show()

# Call the function
plot_satisfaction_by_region(df)
```

```python
# Define a function to create a scatter plot for hours worked and work-life balance
def plot_hours_vs_balance(data):
    """
    Plots a scatter plot showing the relationship between hours worked per week and work-life balance rating.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Hours_Worked_Per_Week', y='Work_Life_Balance_Rating', hue='Work_Location', data=data)
    plt.title("Hours Worked Per Week vs. Work-Life Balance Rating")
    plt.xlabel("Hours Worked Per Week")
    plt.ylabel("Work-Life Balance Rating")
    plt.legend(title="Work Location")
    plt.show()

# Call the function
plot_hours_vs_balance(df)
```

```python
# Define a function to create a heatmap of correlations
def plot_correlation_heatmap(data):
    """
    Plots a heatmap to show correlations between numerical columns.
    """
    plt.figure(figsize=(10, 6))
    numerical_data = data.select_dtypes(include='number')
    correlation_matrix = numerical_data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title("Correlation Heatmap of Numerical Features")
    plt.show()

# Call the function
plot_correlation_heatmap(df)
```
